```python
from pandas import DataFrame, Series
import pandas as pd
import re
from bs4 import BeautifulSoup
from urllib import request
from ast import literal_eval
from datetime import timedelta, datetime as dt
import matplotlib.pyplot as plt
import string
%matplotlib inline
```

```python
top20 = pd.read_pickle('top20movies.pkl')
```

```python
movie_paths = []
for t in list(top20.original_title):
    if t == '8 femmes':
        t = '8 women'
    t = t.lower().replace(' ', '_')
    t = t.lower().replace('.', '')
    movie_paths.append('https://www.rottentomatoes.com/m/' + t)
movie_paths.remove('https://www.rottentomatoes.com/m/사마리아')
movie_paths
```

```python
url = 'https://www.rottentomatoes.com/m/halbe_treppe'
```

```python
ua_header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.67 Safari/537.36'}
req = request.Request(url, headers=ua_header)
```

```python
content = request.urlopen(req)
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
rt_scores = []
for url in movie_paths:
    #print(url[32:])
    try:
        ua_header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.67 Safari/537.36'}
        soup = BeautifulSoup(request.urlopen(request.Request(url, headers=ua_header)), 'html.parser')
        rt_score = soup.findAll('span', class_='meter-value superPageFontColor')[0].text
        #print(rt_score)
        rt_scores.append(rt_score)
    except:
        rt_scores.append('No score found')
rt_scores
```
