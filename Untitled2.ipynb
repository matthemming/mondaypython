```python
#!/usr/bin/env python
# coding: utf-8

# In[4]:


def place_list():
    from urllib import request
    from bs4 import BeautifulSoup
    import pandas as pd
    import re
    
    def pull_down(url):
        req= request.Request(url,headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'})
        contents = request.urlopen(req)
        book = contents.read().decode('utf-8')
        soup = BeautifulSoup(book, 'html.parser')
        
        return soup

    def href_down(url):    
        soup = pull_down(url)   
        href_urls = [] 
        n = soup.find('div', id='mw-content-text')
        for m in n.findAll('a', href=True):
            o = m.get('href')
            if 'List_of_United_Kingdom' in o and 'title' not in o:
                href_urls.append(o)
    
        return href_urls


    href_urls = href_down('https://en.wikipedia.org/wiki/List_of_United_Kingdom_locations')

    d = {}
    test = {}
    out = []
    for h in href_urls:
        hurl = 'https://en.wikipedia.org'+ h
        
        soup = pull_down(hurl)
        for t in soup.findAll('table', class_='wikitable'):
            vcard = t.findAll('tr', class_='vcard')
            try:    
                if len(vcard) == 0:
                    for s in t.findAll('tr'):
                        i = s.get_text()
                        place = i.split('\n')[1]
                        if place != 'Location':
                            locality = i.split('\n')[3]
                            cords = i.split('\n')[5].split('W')[0]+'W'
                            d[place] = [locality,cords]
                            out.append(place) 
                            test[hurl]=place
                else:
                    for s in vcard:
                        place = s.find('a', href = True).get_text()
                        out.append(place)
                        locality = s.find('td', class_='adr').get_text().rstrip("\n")
                        c = s.find('span', class_='geo-dec')
                        if c == None:
                            d[place] = [locality,'N/A']
                        else:
                            cords = c.get_text()
                            d[place] = [locality,cords]
                            test[hurl]=place
            except Exception as e:
                '''Need to print if you want to use dictionary print(hurl)
                print(e)
                print(c)
                print(place)
                print(locality)
                print(cords)'''
                
              
    output = set(out)            
    
    return output


# In[5]:


lst = place_list()


# In[6]:


lst


# In[ ]:




```
