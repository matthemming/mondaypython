```python
import gzip
import re
from datetime import datetime as dt
```

```python
# who is looking at gifs between 1 and 2 am?
```

```python
nasa = r'C:\Data\NASA_access_log_Jul95.gz'
```

```python
count = 0
with gzip.open(nasa) as f:
    for line in f:
        #count +=1
        if count < 10:
            print (line)
            try:
                cleanline = line.decode('utf-8').strip()
                if (re.search('.gif', cleanline)) and (int(cleanline.split(':')[1]) == 1):
                    who = cleanline.split()[0]
                    print('{} viewed a gif between 1 and 2'.format(who))
                    count +=1
            except Exception as e:
                print(e)
#        else:
#            break
            
```
