### Caveats
* Only regestering ports in the portlog. Seem to be a lot missing. e.g. Durban lat long (-29.8697	31.0324	)
    * More of a problem for Q2

* If a ship docked and left in less than 24 hours this would be missed

* Assumes journey data always appears chronologically (seems to be the case)

```python
max_dwt = 60000
port_list = ['PontaDaMadeira', 'Tubarao', 'PontaUbu', 'PortoAcu', 'ItaguaiGuaiba']
```

```python
large_fleet = fleet.loc[fleet.dwt > max_dwt]['imo']
```

```python
vsl_pos_lf = vsl_pos.loc[vsl_pos['imo'].isin(large_fleet)]
```

```python
vsl_pos_ports = vsl_pos_lf
```

```python
vsl_pos_ports['port'] = ''
```

```python
for port in range(0,len(portlog)):
    vsl_pos_ports.loc[(vsl_pos_ports.lat > portlog.iloc[port].latmin)
            & (vsl_pos_ports.lat < portlog.iloc[port].latmax)
            & (vsl_pos_ports.lon > portlog.iloc[port].longmin)
            & (vsl_pos_ports.lon < portlog.iloc[port].longmax), 'port'] = portlog.iloc[port]['name']
    
```

```python
#Should all be in Tubarao
vsl_pos_ports.loc[
    (vsl_pos_ports.lat > -20.4) &
    (vsl_pos_ports.lat < -19.67) &
    (vsl_pos_ports.lon > -40.64) &
    (vsl_pos_ports.lon < -39.64)].head()
```

```python
#Can see journies
vsl_pos_ports.loc[vsl_pos_ports['imo']==9457012]
```

```python
vsl_pos_ports['status'] = ''
```

```python
vsl_pos_ports.head()
```

```python
#Really slow way of doing this.
#If you want to see the result a csv is available in the Google Drive folder I shared with you.
# 'data/all_large_ship_journies.csv' 
#Maybe use apply but unsure at looking at previous/next rows

all_ship_journies = pd.DataFrame()
for ship in large_fleet:
    ship_journey = vsl_pos_ports.loc[vsl_pos_ports['imo']==ship]
    ship_journey.reset_index(inplace=True)
    status = ''
    for i in range(0,len(ship_journey)):
        try:
            #Flags a ship having left if it arrives and leaves in the same day
            #or if it leaves a port and arrives at another the next day
            if (
                ship_journey['port'][i] != ''
                and ship_journey['port'][i] != ship_journey['port'][i-1]
                and ship_journey['port'][i] != ship_journey['port'][i+1]
                ):
            
                ship_journey['status'][i] = 'left'
                status = ''
        
            #Flags a ship as arriving in a port
            elif (
                ship_journey['port'][i] != ''
                and ship_journey['port'][i] != ship_journey['port'][i-1]
                ):
            
                ship_journey['status'][i] =  'arrived'
                status = 'arrived'
            
            #Flags a ship as leaving a port it it has been for multiple days    
            elif (
                status == 'arrived'
                and ship_journey['port'][i] != ship_journey['port'][i+1]
                ):
            
                ship_journey['status'][i] = 'left'
                status = ''
            else:
                continue
        except KeyError:
            continue
    all_ship_journies = all_ship_journies.append(ship_journey)
```

```python
#persist file to save recreating
all_ship_journies.to_csv('data/all_large_ship_journies.csv', index=False)
```

```python
#Pull all ships that have left Brazilian ports.
#Left is only the status if they have also arrived in the time period we are interested in
from_brazil = all_ship_journies.loc[
    (all_ship_journies['port'].isin(port_list)) & (all_ship_journies['status'] == 'left')]
```

```python
#Get dwt for all ships
from_brazil= from_brazil.merge(fleet, how='left', on='imo')
```
