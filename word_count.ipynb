```python
from urllib import request
from pprint import pprint
from bs4 import BeautifulSoup
import re # regular expressions 
import nltk # natural language toolkit
```

```python
url = 'https://en.wikipedia.org/wiki/Botany'
```

```python
req = request.Request(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.67 Safari/537.36'})
contents = request.urlopen(req)
```

```python
botanypage = contents.read().decode('utf-8')
```

```python
soup = BeautifulSoup(botanypage, 'html.parser')
```

```python
soup.get_text()
```

```python
mystring = r'You can use beautiful soup to parse html; a can of regular soup can be used to eat for dinner.'
```

```python
words = mystring.split()
```

```python
words
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
stop = nltk.corpus.stopwords.words('english')
```

```python
ps = PorterStemmer()
```

```python
clean_words = []
for w in words:
    w = w.lower()
    w = re.sub('[^a-zA-Z]+', '', w)
    w = ps.stem(w)
    if w not in stop:
        clean_words.append(w)
```

```python
clean_words
```

```python
# Word Count

d = {}
for w in clean_words:
    d[w] = d.get(w, 0) + 1
```

```python
sorted(d.items(), key = lambda x: x[1], reverse=True)
```
