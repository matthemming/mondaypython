```python
from bs4 import BeautifulSoup
from urllib import request
from pprint import pprint
import pandas as pd
```

Norfolk fire service puts out a live incident feed for the last 10 days

The code blow creates a list of incident dictionaries containing the event id, date, time, location and description 

```python
url = 'http://live.norfolkfireservice.co.uk/wp/'
req = request.Request(url, headers={'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.67 Safari/537.36'})
content = request.urlopen(req)
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
table_content = soup.find('table', class_='mrAntGrid').find('tbody')

incids_list = []

for row in table_content.findAll('tr'):
    #row.findAll('td')
    row_dict = {} 
    row_dict['id'] = row.findAll('td')[0].text
    row_dict['date'] = row.findAll('td')[1].text.replace(u'\xa0', u' ')[:-8]
    row_dict['time'] = row.findAll('td')[1].text.replace(u'\xa0', u' ')[-8:]
    row_dict['location'] = row.findAll('td')[2].text
    row_dict['text'] = row.findAll('td')[4].text
    
    incids_list.append(row_dict)

incids_list
```

```python
pd.DataFrame(incids_list).to_csv(r'C:\git\fire\data\Fire_Regions\NorfolkFire.csv')
```
