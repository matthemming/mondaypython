```python
from bs4 import BeautifulSoup
import re
from urllib import request
from pprint import pprint
from datetime import datetime as dt
```

```python
url = 'https://www.theguardian.com/football/premierleague/results'
```

```python
req = request.Request(url, headers={'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.67 Safari/537.36'})
```

```python
content = request.urlopen(req)
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
matchday_sections = soup.find_all('div', class_= 'football-matches__day')
```

```python
test_date = matchday_sections[0].find('div', class_='date-divider').string
test_date
```

```python
dt.strptime(test_date, '%A %d %B %Y').strftime('%Y-%m-%d')
```

```python
matchdays = {} # dict of date:[all match results in a list of dicts]

for i in soup.find_all('div', class_= 'football-matches__day'):
    day_table = i.find('table', class_='table')

    #single_matches = day_table.findAll('tr', class_='football-match')[0]
    date = i.find('div', class_='date-divider').string
    date = dt.strptime(date, '%A %d %B %Y').strftime('%Y-%m-%d')
    scores_from_day = []
    for match in day_table.findAll('tr', class_='football-match'):
        #the below is for a single match
        teams = match.find('td', class_='football-match__teams')
        match_result = {}
        teamslist = []
        for team in teams.findAll('div', class_='football-team__name'):
            teamslist.append(team.find('span', class_='team-name__long').string)
        
        score = []
        for goals in teams.findAll('div', class_='football-team__score'):
            score.append(int(goals.string))
        
        for i in range(len(teamslist)):
            match_result[teamslist[i]] = score[i]
        scores_from_day.append(match_result)
        
    matchdays[date] = scores_from_day 
```
