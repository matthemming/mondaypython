```python
url = 'https://www.bbc.co.uk/news'
```

```python
req = request.Request(url, headers={'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.67 Safari/537.36'})
```

```python
content = request.urlopen(req)
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
#nw-o-no-keyline is a class identified while inspecting the web page that appears in headings

headlines = soup.find_all('div', class_= 'nw-o-no-keyline')
```

```python
for h in headlines:
    print(h.get_text())
```
